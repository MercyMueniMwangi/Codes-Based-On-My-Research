```
# This program predicts stock prices by using machine learning models

#import packages
import quandl
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

**Importing Data**

```
# data
dataset =quandl.get("WIKI/AMZN")
dataset.head()
```

```
dataset.dtypes
```

```
dataset.describe()
```

```
dataset.shape
```

```
dataset.isna().sum()
```

**Defining Variables**

```
#get the adjusted close price ad independent variable
dataset =dataset[["Adj. Close"]]
dataset
```

```
# variable for predicting n days to the future say the next day
forecast_out = 30
#Create another column (the target or dependent variable) shifted 'n' units up
dataset['Prediction'] = dataset[['Adj. Close']].shift(-forecast_out)
dataset.head(30)
```

```
dataset.isna().sum() 
```

**Creating Variables**

```
#Create the independent data set (X)
# Then convert the dataframe to a numpy array
X = np.array(dataset.drop(['Prediction'],1))
#remove the last n rows
X =   X[: - forecast_out]
print(X)
```

```
#Create the independed dataset Y
# here we convert the data frame to a numpy array (All of the values including the NaN's)
Y = np.array(dataset['Prediction'])
# remove NaN
Y=Y[:-forecast_out]
print(Y)
```

**Spliting data**

```
trainx,testx,trainy,testy=train_test_split(X,Y,test_size=0.25)
```

**Fitting the model**


Support vector machine(rergressor

```
svm_regressor=SVR(kernel='rbf', C=1e3, gamma=0.1)
svm_regressor.fit(trainx, trainy)
```

**Scoring the model**

```
#will use r-squared to score the model
# The best possible score is 1.0
svm_score=svm_regressor.score(testx,testy)
print("svm confidence Score: ", svm_score)
```

```
linear_regressor = LinearRegression()
# Train the model
linear_regressor.fit(trainx, trainy)
```

```
# Score using the coefficient of determination R^2 of the prediction. 
# The best possible score is 1.0
linear_regressor_confidence = linear_regressor.score(testx,testy)
print("linear_regressor_Score: ", linear_regressor_confidence)
```

**Predicting future stock values**

```
# Set x_forecast equal to the last 30 rows of the original data set from Adj. Close column
x_forecast = np.array(dataset.drop(['Prediction'],1))[-forecast_out:]
print(x_forecast)
```

```
# Print linear regression model predictions for the next 'n' days
linear_prediction = linear_regressor.predict(x_forecast)
print(linear_prediction)


```

```
# Print support vector regressor model predictions for the next 'n' days
svm_prediction = svm_regressor.predict(x_forecast)
print(svm_prediction)
```
